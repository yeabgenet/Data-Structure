```python
def maximum_saving(input_network):

    n = 7  # number of nodes
    edges = []
    total_weight = 0

    # Convert adjacency matrix to edge list
    for i in range(n):
        for j in range(i+1, n):  # only upper triangle to avoid duplicates
            val = input_network[i][j]
            if val != '-':
                w = int(val)
                edges.append((i+1, j+1, w))  # 1-indexed
                total_weight += w

    # ---- Kruskal's Algorithm ----
    class DSU:
        def __init__(self, n):
            self.parent = list(range(n))
            self.rank = [0]*n

        def find(self, x):
            if self.parent[x] != x:
                self.parent[x] = self.find(self.parent[x])
            return self.parent[x]

        def unite(self, a, b):
            a = self.find(a)
            b = self.find(b)
            if a == b:
                return False
            if self.rank[a] < self.rank[b]:
                a, b = b, a
            self.parent[b] = a
            if self.rank[a] == self.rank[b]:
                self.rank[a] += 1
            return True

    edges_sorted = sorted((w, u-1, v-1) for u, v, w in edges)
    dsu = DSU(n)
    mst_weight = 0
    for w, u, v in edges_sorted:
        if dsu.unite(u, v):
            mst_weight += w

    return total_weight - mst_weight
```
